# Experiments — Brain Bling (RACE RC Project)
Tracks model experiments, hyperparameter tuning, and result comparison.

## 1. Setup

```python
import pandas as pd
import numpy as np
import pickle
import os
import matplotlib.pyplot as plt
import seaborn as sns

MODELS_A = '../models/model_a/traditional'
MODELS_B = '../models/model_b/traditional'
print('Setup complete')
```

## 2. Model A — Binary Classifier Results

```python
binary_df = pd.read_csv(f'{MODELS_A}/binary_classification_metrics.csv')
option_df = pd.read_csv(f'{MODELS_A}/option_selection_metrics.csv')

print('Binary Classification Metrics:')
display(binary_df)
print('\nOption Selection (Exact Match):')
display(option_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(binary_df['Model'], binary_df['Macro F1'], color='steelblue')
axes[0].set_title('Model A — Macro F1')
axes[0].set_ylabel('Macro F1')
axes[0].tick_params(axis='x', rotation=20)

axes[1].bar(option_df['Model'], option_df['Exact Match'], color='darkorange')
axes[1].set_title('Model A — Exact Match (Option Accuracy)')
axes[1].set_ylabel('Exact Match')
axes[1].tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.show()
```

## 3. Model A — Ensemble Results

```python
ens_bin = pd.read_csv(f'{MODELS_A}/ensemble_binary_metrics.csv')
ens_opt = pd.read_csv(f'{MODELS_A}/ensemble_option_metrics.csv')

display(ens_opt[['Model', 'Option Accuracy', 'Option Macro F1']])
```

```python
plt.figure(figsize=(10, 5))
plt.bar(ens_opt['Model'], ens_opt['Option Accuracy'], color='mediumseagreen')
plt.title('Ensemble — Option Accuracy (Exact Match)')
plt.ylabel('Option Accuracy')
plt.axhline(0.25, linestyle='--', color='red', label='Random baseline (25%)')
plt.legend()
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

## 4. Model A — Unsupervised / Semi-Supervised

```python
unsup_df = pd.read_csv(f'{MODELS_A}/phase4_comparison.csv')
display(unsup_df)
```

## 5. Model B — Distractor Results

```python
dist_df = pd.read_csv(f'{MODELS_B}/distractor_results.csv')
print(f'Samples evaluated: {len(dist_df)}')
print(dist_df[['ml_precision','ml_recall','ml_f1']].describe())

plt.figure(figsize=(8,4))
plt.hist(dist_df['ml_f1'], bins=20, color='mediumpurple', edgecolor='black')
plt.title('Distractor F1 Distribution')
plt.xlabel('Token F1')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 6. Model B — Hint Results

```python
hint_df = pd.read_csv(f'{MODELS_B}/hint_results.csv')
display(hint_df)
```

## 7. Overall Model Comparison

```python
summary = {
    'Component': [
        'LR (One-Hot)', 'SVM (One-Hot)', 'NB (BoW)', 'RF (Lexical)',
        'Soft Voting', 'Hard Voting', 'Stacking'
    ],
    'Exact Match': [
        option_df.loc[option_df.Model=='Logistic Regression','Exact Match'].values[0],
        option_df.loc[option_df.Model=='SVM','Exact Match'].values[0],
        option_df.loc[option_df.Model=='Naive Bayes','Exact Match'].values[0],
        option_df.loc[option_df.Model=='Random Forest','Exact Match'].values[0],
        ens_opt.loc[ens_opt.Model=='Soft Voting','Option Accuracy'].values[0],
        ens_opt.loc[ens_opt.Model=='Hard Voting','Option Accuracy'].values[0],
        ens_opt.loc[ens_opt.Model=='Stacking','Option Accuracy'].values[0],
    ]
}
summary_df = pd.DataFrame(summary)

plt.figure(figsize=(12, 5))
colors = ['steelblue']*4 + ['darkorange']*3
plt.bar(summary_df['Component'], summary_df['Exact Match'], color=colors)
plt.axhline(0.25, linestyle='--', color='red', label='Random baseline (25%)')
plt.title('Model A — Exact Match Comparison (All Models)')
plt.ylabel('Exact Match')
plt.legend()
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()

display(summary_df)
```
